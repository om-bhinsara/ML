```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
## create the dataset
X, y = make_classification(n_samples=1000, n_features=10, n_classes=2, random_state=15)
```

```python
# standardized values - no need to do standardization because currently i am focusing on implementation only
pd.DataFrame(X)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# import model  
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()

# model train 
logistic.fit(X_train,y_train)

# do prediction on test data
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
# check performance matrices
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
score=accuracy_score(y_pred,y_test)
print("socre is ", score)

print(classification_report(y_pred,y_test))

print("Confusion matrix is ")
print(confusion_matrix(y_pred,y_test))
```

# Hyperparameter Tunning and Cross Validation 

## Grid SearchCv

```python
param_grid = [
    {
        'solver': ['liblinear'],
        'penalty': ['l1', 'l2'],
        'C': [100, 10, 1.0, 0.1, 0.01]
    },
    {
        'solver': ['lbfgs', 'newton-cg', 'sag'],
        'penalty': ['l2'],
        'C': [100, 10, 1.0, 0.1, 0.01]
    },
    {
        'solver': ['saga'],
        'penalty': ['l1', 'l2', 'elasticnet'],
        'C': [100, 10, 1.0, 0.1, 0.01],
        'l1_ratio': [0.5]  # required for elasticnet
    }
]
```

## C Value in Machine Learning (Logistic Regression / SVM)

The **C value** is a **regularization parameter** used in models like **Logistic Regression** and **Support Vector Machines (SVM)**.

### Meaning of C
- **C controls the trade-off** between:
  - fitting the training data well  
  - keeping the model simple (regularized)

It is the **inverse of regularization strength**.

### Effect of C Value
- **Small C (e.g., 0.01, 0.1)**  
  → Strong regularization  
  → Simpler model  
  → May underfit  

- **Large C (e.g., 10, 100)**  
  → Weak regularization  
  → More complex model  
  → May overfit  

### Why C is Tuned in GridSearchCV
Different C values can significantly change model performance.  
GridSearchCV tests multiple C values to find the best balance between **bias and variance**.

### One-Line Definition
**C is a regularization parameter that controls how strictly a model fits the training data.**


## Solvers in GridSearchCV

In GridSearchCV, a **solver** is not a component of GridSearchCV itself, but a **hyperparameter of the machine learning model**. It defines the **optimization algorithm** used to minimize the loss function and learn the model parameters.

### Role of Solvers
Solvers control how efficiently and accurately a model is trained. The choice of solver affects:
- Training speed  
- Convergence behavior  
- Type of regularization supported  
- Suitability for small or large datasets  

### Example: Logistic Regression Solvers
- **liblinear** – Suitable for small datasets; supports L1 and L2 regularization  
- **lbfgs** – Efficient and commonly used; supports multiclass classification  
- **newton-cg** – Uses second-order optimization; good for larger datasets  
- **sag** – Fast for large datasets; requires feature scaling  
- **saga** – Supports L1, L2, and ElasticNet regularization; scalable  

### Use in GridSearchCV
GridSearchCV evaluates different solvers by treating them as hyperparameters and selects the solver that yields the best performance based on the chosen scoring metric.

### Summary
Choosing the right solver helps achieve faster convergence, better accuracy, and stable model training.


```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

## Stratified K-Fold Cross-Validation

Stratified K-Fold Cross-Validation is used mainly for **classification problems**.  
It splits the data into K folds such that **each fold maintains the same class distribution** as the original dataset.

This method is especially useful for **imbalanced datasets**, as it ensures that all classes are represented in every fold, leading to **more reliable and unbiased model evaluation**.


```python
## GridSearchCV
from sklearn.model_selection import GridSearchCV
model=LogisticRegression()
grid=GridSearchCV(estimator=model,param_grid=param_grid,scoring='accuracy',cv=cv,n_jobs=-1)

''' 
This line applies GridSearchCV to find the best hyperparameters for a given model.  
It tests all parameter combinations in `param_grid`, evaluates them using **accuracy**, applies 
**cross-validation (cv)**, and uses **all CPU cores** (`n_jobs = -1`) to speed up the process.
'''
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)

# classification report will pritn precesion, recall, and f1-score 
print(classification_report(y_pred,y_test))

# pritn confusion matrix
print(confusion_matrix(y_pred,y_test))
```

## Randomized SearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist = [
    {
        'solver': ['liblinear'],
        'penalty': ['l1', 'l2'],
        'C': [100, 10, 1.0, 0.1, 0.01]
    },
    {
        'solver': ['lbfgs', 'newton-cg', 'sag'],
        'penalty': ['l2'],
        'C': [100, 10, 1.0, 0.1, 0.01]
    },
    {
        'solver': ['saga'],
        'penalty': ['elasticnet'],
        'C': [100, 10, 1.0, 0.1, 0.01],
        'l1_ratio': [0.1, 0.5, 0.9]
    }
]
```

```python
model=LogisticRegression()
randomcv=RandomizedSearchCV(estimator=model, param_distributions=param_dist, cv=5, scoring='accuracy')
```

```python
randomcv.fit(X_train,y_train)
```

```python
randomcv.best_score_
```

```python
randomcv.best_params_
```

```python
y_pred=randomcv.predict(X_test)
```

```python
score=accuracy_score(y_pred,y_test)
print(score)

print(classification_report(y_pred,y_test))

print(confusion_matrix(y_pred,y_test))
```
